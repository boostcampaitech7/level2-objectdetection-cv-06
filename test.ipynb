```python
import json
import random

# Load train.json
with open('../../dataset/train.json') as f:
    train_data = json.load(f)

# Randomly select a portion of the data for validation
val_data = {
    'images': [],
    'annotations': [],
    'categories': train_data['categories']  # Assuming categories are the same
}

# Choose 10% of the training data for validation
num_val_samples = int(len(train_data['images']) * 0.2)
val_images = random.sample(train_data['images'], num_val_samples)

for img in val_images:
    val_data['images'].append(img)
    # Add related annotations
    img_id = img['id']
    val_data['annotations'].extend(
        [ann for ann in train_data['annotations'] if ann['image_id'] == img_id]
    )

# Save to val.json
with open('../../dataset/val.json', 'w') as f:
    json.dump(val_data, f)
```
